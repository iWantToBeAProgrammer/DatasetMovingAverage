```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_excel('/content/drive/My Drive/Colab Notebooks/DataMining/hargasaham-training-small.xls')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
# Mengkonversi Kolom Tanggal
```

```python
# Menghitung 50-Day dan 200-Day Moving Average
data['MA_50'] = data['Close'].rolling(window=50).mean()
data['MA_200'] = data['Close'].rolling(window=200).mean()
```

```python
# Plot harga penutupan dengan 50-day dan 200-day moving average
plt.figure(figsize=(12, 6))

# Plot harga penutupan
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')

# Plot moving average 50 hari
plt.plot(data['Date'], data['MA_50'], label='50-Day MA', color='orange')

# Plot moving average 200 hari
plt.plot(data['Date'], data['MA_200'], label='200-Day MA', color='green')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Closing Price with 50-Day & 200-Day Moving Average')
plt.legend()
plt.show()
```
